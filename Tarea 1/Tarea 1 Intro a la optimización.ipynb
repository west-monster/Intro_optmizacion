## Punto 1

Si en la matriz $A$ todas sus columnas son linealmnete independientes entonces $A^T A$ tiene inversa. 

Lema: 
$$ rango(A) = rango(A^TA)$$ 

Demostración (Contradicción):

Supongamos que todas las columnas de $A$ son linealmente independientes y que $A^T A$ no tiene inversa. Sabemos que $A$ es una matriz $mxn$ donde $m \geq n$ para que el problema tenga sentido, entonces el $rango(A) = n$, pues $n\leq m$ y todas las columnas son linealmnete independientes. Por otro lado tenemos que $A^T A$ es una matriz $nxn$ y usando el lema anterior tenemos que $rango(A^TA) = n$ por lo que podemos concluir que $A^T A$ es de rango completo =><= pues $A^T A$ no tiene inversa. 


Por otro lado, un sistema $Cx = b$ tiene solución si y solo si $b$ es una combinación lineal de las columnas de A, por lo que en este caso seria necesario que $b$ sea combinación lineal de las columnas de $A^T A$

Para este punto usamos los datos del precio del Bitcoin desde el 17 de marzo del 2021 hasta el 17 de marzo del 2022. Usaremos los precios que se registraron al momento del cierre del mercado. Note que en este caso tambien es necesario invertir el orden de los precios, ya que en el dataframe el primer dato no corresponde al precio del 17 de marzo del 2021 sino al precio del 17 de marzo del 2022.

Usaremos el metodo de minimos cuadrados para modelos no lineales. En este caso vamos a usar funciones polinomicas de grado 2 hasta 5 e iremos viendo cómo se ajustan a la gráfica. Usaremos 366 datos, por lo que nuestro polinomio será evaluado en los días 1 al 366 en la gráfica. 

```julia
N = size(dataBitcoin.PriceFloat,1) #Guardamos la cantidad de datos que tenemos.

days = Array(1:N) #Creamos un array con los dias con los que vamos a estimar el polinomio.

Polyvalues=[] #Guardamoes en este array los coeficientes de los polinomios
 
b = dataBitcoin.PriceFloat; #Definimos el vector b con los datos del precio del bitcoin.

Polydata =  [ones(N,1) ones(N,1)] #creamos un array para guardar los datos de nustro polinomio

#Bucle para usar el metodo de minimos cuadrados con polinomios de grado 2 hasta 5
for degree in 2:5

    #Creamos la matriz A con los dalores de days 
    A = [ones(N,1) days]
    
    #Si el grado de nuestro polinomio es mayor que 2, ejecutamos el bucle 
    if degree >= 2
        for i = 2:degree
          A = [A days.^i]  
        end   
    end 
    
    #Resolvemos el sistema y guardamos los coeficientes del polinomio para usarlos posteriormnete 
    xhat = A\b
    append!(Polyvalues,[xhat])
         
    
    poly = xhat[1]*ones(N,1)
       
    #Evaluamos cada polinomio desde el día 1 hasta el 366
    for i = 2:degree + 1
        poly = poly + xhat[i]*A[:,i]  
    end 
    
    #Guardamos en el Array Polydata la información anterior
    Polydata = [Polydata poly]  
    
end 


#Creamos el plot y renomabramos las funciones
p = PlotlyJS.plot(days, [b Polydata[:,3:6]]);
restyle!(p, name=["Bitcoin", "2 Grade Poly", "3 Grade Poly", "4 Grade Poly", "5 Grade Poly"]);


p
```

En la gráfica anterior podemos ver que el polinomio de grado 5 parece ajustarse mejor a los datos, razón por la cual vamos a usar este modelo.


Ahora comparamos los precios del Bitcoin de esta semana, para ello cargaremos los datos de los últimos 7 días 
(estos datos son posteriores a los que se utilizaron en el modelo y pertenecen a otro CSV). En este caso tambien es necesario invertir el orden del estos datos debido a lo que se explicó al inicio. 

```julia
#Datos del valor del Bitcoin desde el 18 de marzo del 2022 hasta el 24 de marzo del 2022
dataBitcoinActual = CSV.read("BTC_actual.csv",DataFrame,delim=',')
dataBitcoinActual.Price .= replace.(dataBitcoinActual.Price, "," => "");
dataBitcoinActual.PriceFloat = reverse(parse.(Float64, dataBitcoinActual.Price));
dataBitcoinActual.PriceFloat
```

Ahora usamos los coeficientes del polinomoio de grado 5 que se encuentran en la matriz Polydata (que almacena los coeficientes del modelo anterior de los polinomios de grado 2 hasta 5) y evaluamos en los días 367 al 373.

```julia
#Creamos un vector con los números del 367 al 373, estos serán los días en los cuales evaluaremos el polinomio. 
days_p = 367:373

#Usamos los coefientes del polinomio de grado 5, que están guardados en la posición 4 del array
Grade_five_Poly_data = Polyvalues[4][6]*days_p.^5 + Polyvalues[4][5]*days_p.^4 + Polyvalues[4][4]*days_p.^3 + 
Polyvalues[4][3]*days_p.^2 + Polyvalues[4][2]*days_p + Polyvalues[4][1]*ones(7,1)
```

Ahora le restamos el vector con los valores reales al vector con los valores predichos para los dias 367 al 373.

```julia
error_pred = dataBitcoinActual.PriceFloat - Grade_five_Poly_data
```

En este caso la máxima diferencia entre el valor real y el predicho es de -11700.6499 dolares.

## Punto 3

Cargamos los precios del oro, la plata y el petroleo, en este caso como los precios de las commodities no se registran los fines de semana y los del Bitcoin si, por lo que fue necesario elimanar los datos de los días en los cuales se registró el precio del Bitcoin pero no se registró el precio de estas commodities.

```julia
#Datos del valor del Bitcoin desde el 17 de marzo del 2021 hasta el 17 de marzo del 2022
dataBitcoin2 = CSV.read("BTC.csv",DataFrame,delim=',')
dataBitcoin2.Price .= replace.(dataBitcoin2.Price, "," => "");
dataBitcoin2.PriceFloat = parse.(Float64, dataBitcoin2.Price)


#Cargamos los datos del Oro desde el 17 de marzo del 2021 hasta el 17 de marzo del 2022
dataGold = CSV.read("Gold.csv",DataFrame,delim=',')
dataGold.Price .= replace.(dataGold.Price, "," => "");
dataGold.PriceFloat = parse.(Float64, dataGold.Price);

#Cargamos los datos de la plata desde el 17 de marzo del 2021 hasta el 17 de marzo del 2022
dataSilver = CSV.read("Silver.csv",DataFrame,delim=',')
rename!(dataSilver, :Price => :PriceFloat) #estos datos ya son flotantes, por lo que renombramos la columna 

#Cargamos los datos del Petroleo desde el 17 de marzo del 2021 hasta el 17 de marzo del 2022
dataOil = CSV.read("Brent_Oil.csv",DataFrame,delim=',');
rename!(dataOil, :Price => :PriceFloat) #estos datos ya son flotantes, por lo que renombramos la columna 

inter = intersect(dataBitcoin2.Date, dataOil.Date, dataSilver.Date, dataGold.Date);

#Los precios de las comodities no se registran los fines de semana, pero los del bitcoin si.
#Se procede a filtar por las fechas que coincden en los 4 data frames

dataOil = dataOil[in(inter).(dataOil.Date), :];
dataSilver = dataSilver[in(inter).(dataSilver.Date), :];
dataGold = dataGold[in(inter).(dataGold.Date), :];
dataBitcoin2 = dataBitcoin2[in(inter).(dataBitcoin2.Date), :];

dataBitcoin2
```

Ahora que los data frames están listos, procedemos a utilizar el Metodo de minimos cuadrados para relacionar el precio del Bitcoin con el precio del Oro, la Plata y el petroleo (de referencia brent). 

```julia
#Guardamos la cantidad de datos que tenemos.
N = size(dataBitcoin2.PriceFloat,1) 

#Definimos el vector b con los datos del precio del bitcoin.
b = dataBitcoin2.PriceFloat;

#Construimos la matriz de diseño
A = [ones(N,1) dataGold.PriceFloat dataOil.PriceFloat dataSilver.PriceFloat];

xhat = A\b 
```

Ahora probamos el modelo con algunos datos (los del 17 de marzo del 2022), para este fecha el precio del Bitcoin fue: 40910.0

```julia
#Creamos una función con los pesos anteriores
relation(Gold, Oil, Silver) = Gold*xhat[2] + Oil*xhat[3] + Silver*xhat[4] + xhat[1]

res = relation(1943.20 ,106.64, 25.616)
```

Ahora usamos nuestro modelo con los datos de las commodities de la semana actual (del 18 de marzo del 2022 al 24 de marzo del 2022) y comparamos con el precio del Bitcoin en el mismo lapso de tiempo. 

```julia
#Cargamos los datos del Oro de esta semana
dataGoldActual = CSV.read("Gold_actual.csv",DataFrame,delim=',')
dataGoldActual.Price .= replace.(dataGoldActual.Price, "," => "");
dataGoldActual.PriceFloat = parse.(Float64, dataGoldActual.Price);

#Cargamos los datos de la plata de esta semana
dataSilverActual = CSV.read("Silver_actual.csv",DataFrame,delim=',')
rename!(dataSilverActual, :Price => :PriceFloat);

#Cargamos los datos del Petroleo de esta semana
dataOilActual = CSV.read("Brent_oil_actual.csv",DataFrame,delim=',');
rename!(dataOilActual, :Price => :PriceFloat);

#Cargamos los datos del Bitcoin de esta semana
dataBitcoinActual2 = CSV.read("BTC_actual.csv",DataFrame,delim=',');
dataBitcoinActual2.Price .= replace.(dataBitcoinActual2.Price, "," => "");
dataBitcoinActual2.PriceFloat = parse.(Float64, dataBitcoinActual2.Price);



inter2 = intersect(dataBitcoinActual2.Date, dataOilActual.Date, dataSilverActual.Date, dataGoldActual.Date);
dataOilActual = dataOilActual[in(inter2).(dataOilActual.Date), :];
dataGoldActual = dataGoldActual[in(inter2).(dataGoldActual.Date), :];
dataSilverActual = dataSilverActual[in(inter2).(dataSilverActual.Date), :];
dataBitcoinActual2 = dataBitcoinActual2[in(inter2).(dataBitcoinActual2.Date), :];

```

Utilzamos el modelo anterior con los valores actuales del Oro, la plata y el petroleo (del 18 de marzo de 2022 al 24 de marzo del 2022) para ver qué tan cerca está nuestro modelo del valor real del bitcoin

```julia
relatio_next_five_days = dataGoldActual.PriceFloat*xhat[2] + dataOilActual.PriceFloat*xhat[3] + 
dataSilverActual.PriceFloat*xhat[4] + xhat[1]*ones(5,1)
```

Calculamos la diferencia entre el valor estimado y el valor real del Bitcoin
